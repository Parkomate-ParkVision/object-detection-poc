```python
import os
import shutil

source_dir = '/content/runs'
dest_dir = '/content/drive/MyDrive/Colab related files/ANPR POC Data'

for dirpath, dirnames, filenames in os.walk(source_dir):
    relative_dir = os.path.relpath(dirpath, source_dir)
    dest_subdir = os.path.join(dest_dir, relative_dir)
    os.makedirs(dest_subdir, exist_ok=True)
    for filename in filenames:
        src_file = os.path.join(dirpath, filename)
        dest_file = os.path.join(dest_subdir, filename)
        shutil.copy(src_file, dest_file)
```
